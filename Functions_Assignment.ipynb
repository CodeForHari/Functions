Q1. What is the difference between a function and a method in Python?
  - Difference Between Function and Method in Python
    - **Functions**
    
    A function is a block of reusable code that is defined independently and can be called by its name.
    
    1. It is not tied to any object or class.
    2. Created using the **def** keyword.

  - Example:

```python
def add(a, b):
    return a + b

print(add(5, 3))   # Calling a function
```

- **Method**
  - A method is also a function, but it belongs to an object or a class.
  1. It is called using an object.
  2. Defined inside a class.

- Example:

```python
class Calculator:
    def add(self, a, b):  # This is a method
        return a + b

obj = Calculator()
print(obj.add(5, 3))    # Calling a method using an object
```

Q6. Explain the concept of generators in Python and how they are defined.
- **What Are Generators?**
  - Generators are special functions in Python that produce a sequence of values one at a time, instead of returning all values at once.
They are memory-efficient because they generate values only when needed.

- **Generators are used when:**
  - You want to work with large data,
  - Or you don’t want to store the entire result in memory.

- **How Generators Are Defined**
  - Generators are defined just like normal functions, but instead of return, they use the yield keyword.

- **Example of a Generator Function**

```python
def countdown(n):
    while n > 0:
        yield n    # yields one value at a time
        n -= 1
```

- Using the Generator

```python
for number in countdown(5):
    print(number)
```

Q7. What are the advantages of using generators over regular functions?
- **Advantages of Generators Over Regular Functions**
  - Generators have several benefits, especially when working with large data or long sequences.

**1. Memory Efficient**
  - A regular function returns all values at once, storing them in memory.
  - A generator produces one value at a time, so it uses very little memory.

**2. Faster in Large Iterations**
- Generators compute values only when needed (lazy evaluation), which makes them faster for big loops compared to building a large list first.

**3. Infinite Sequences Possible**
- Generators can create sequences that never end — something impossible with normal functions.

**4. Cleaner Code**
  - Generators make code simpler and more readable when dealing with sequences.

**Simple Example Comparing Both**
- Regular Function:

```python
def squares_list(n):
    result = []
    for i in range(n):
        result.append(i*i)
    return result

print(squares_list(5))

# Creates and stores the whole list in memory.
```

- Generator Function:

```python
def squares_gen(n):
    for i in range(n):
        yield i*i

gen = squares_gen(5)
print(next(gen))
print(next(gen))

# Produces values one-by-one without storing the entire list.
```

It is created using the keyword lambda and is used for short, simple operations.

Syntax:
> lambda arguments: expression

  - Lambda functions are commonly used when:

Q9. Explain the purpose and usage of the `map()` function in Python.
- Purpose and Usage of the `map()` Function in Python
  - The `map()` function is used to apply a function to each item of an iterable (like list, tuple, etc.) and return a new iterator with the results.

Syntax:
> map(function, iterable)

**Example 1: Using map() with a normal function**

```python
numbers = [1, 2, 3, 4]
result = map(square, numbers)

print(list(result))
```

**Example 2: Using map() with a lambda function**

```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)

print(list(result))
```

**Example 3: Mapping multiple iterables**

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x + y, a, b)
print(list(result))
```

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- These three functions are used for functional-style programming, but each has a different purpose.

1. `map()` – Transform Each Element
- Applies a function to every element of an iterable and returns a new iterator with the results.

**Example:**


```python
numbers = [1, 2, 3]
result = map(lambda x: x * 2, numbers)
print(list(result))

## Used for modifying or transforming items.
```

2. `filter()` – Select Certain Elements
- Keeps only those elements for which the function returns True.

**Example:**

```python
numbers = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))

## Used for filtering / selecting items.
```

3. `reduce()` – Combine All Elements into One Result
- Repeatedly applies a function to the elements and reduces the iterable to a single value.
(It is in the functools module.)

**Example:**

```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)

## Used for aggregation like sum, product, max, etc.
```

Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:   # check even number
            total = (total+num)
    return total
```

```python
nums = [10, 15, 22, 33, 40]
print(sum_even_numbers(nums))
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
print("Prime numbers from 1 to 200:")
for number in range(1, 201):
    if is_prime(number):
        print(number, end=" ")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fib(n):
  a = 0
  b = 1
  for i in range(n):
    yield a
    a, b = b, a+b
```

```python
fib_sequence = fib(10)
for num in fib_sequence:
    print(num, end=" ")
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(x):
    for i in range(x + 1):
        yield 2**i
```

```python
for p in powers_of_two(5):
    print(p, end=" ")
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line
```

```python
for line in read_file_lines("myfile.txt"):
    print(line)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = filter(lambda ch: ch not in vowels, text)
    return "".join(result)
```

```python
string = "Hello World"
print(remove_vowels(string))
```
